```python
from sklearn.datasets import load_boston
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data_scientist_intern_revenue_model - working file.csv")
data
```

```python
data1 = data[data['Revenue'].notna()]
data1
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data1['NAME_LE'] = le.fit_transform(data1['NAME'])
data1
```

```python
data1.isnull().sum()
```

```python
# df["value"] = df.groupby("name").transform(lambda x: x.fillna(x.mean()))
data1["Employees"] = data1.groupby("NAME").transform(lambda x: x.fillna(x.mean()))
data1
```

```python
data1.info()
```

```python
data1['Revenue'] = data1['Revenue'].replace(',','', regex=True)
data1['Revenue'].astype(str).astype('int64')
data1
```

```python
data1.isnull().sum()
```

```python
X = data1[['NAME_LE','Year','Employees']]
Y = data1[['Revenue']]
Y = Y.reset_index(drop = True)
```

```python
rf = RandomForestRegressor(n_estimators=500)
rf_model = rf.fit(X, Y)
rf_model
```

```python
y_pred = rf_model.predict(X)
y_pred
```

```python
import sklearn.metrics
import math
mse = sklearn.metrics.mean_squared_error(Y,y_pred)
# Root mean Squared Error
rmse = math.sqrt(mse)
rmse
```
